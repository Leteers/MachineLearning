<H1>Кластеризация данных с использованием алгоритма K-средних

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
```

<H3>Загрузка набор данных "Iris" 

```python
data = pd.read_csv("iris.csv")
data1 = data.iloc[:, 0:2]
data2 = data.iloc[:, 2:4]
varietys = ['Setosa','Versicolor','Virginica']
print(varietys.index('Setosa'))
y = [varietys.index(str(data["variety"][i])) for i in range(len(data['variety']))]

print(y)
scaler = MinMaxScaler()
scaler.fit(data1)
data1 = pd.DataFrame(columns = ["sepal.length","sepal.width"] , data = scaler.transform(data1))
scaler.fit(data2)
data2= pd.DataFrame(columns= ["petal.length","petal.width"], data = scaler.transform(data2))

print(data1)
print(data2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data1, y, test_size=0.3)
X_train_1, X_test_1, y_train_1, y_test1 = train_test_split(data2, y,test_size= 0.3)
model = KMeans(n_clusters=3)
model.fit(X_train,y_train)
model1 = KMeans(n_clusters=3)
model1.fit(X_train_1, y_train_1)
```

```python
y_pred = model.predict(X_test)
y_pred1 = model1.predict(X_test_1)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy1 = accuracy_score(y_test1, y_pred1)
print("Accuracy:", accuracy)
print("Accuracy2:", accuracy1)
```
